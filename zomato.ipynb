```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#creating a dataframe
dataframe = pd.read_csv("Zomato data .csv")
print(dataframe)
```

```python
dataframe
```

```python
 #Convert the data type of column rate
```

```python
def handleRate(value):
    value = str(value).split('/')  
    value = value[0]
    return float(value)

dataframe['rate'] = dataframe['rate'].apply(handleRate)
print(dataframe.head())
```

```python
dataframe.info()
```

```python
#TYPE OF RESTURANT
```

```python
dataframe.head()
```

```python
sns.countplot(x='listed_in(type)', data=dataframe, color='red')
plt.xlabel("type of restaurant")
plt.show()


```

```python
#CONCLUSION MAJORITY OF THE RESTAURANTS FALLS UNDER DINING CATEGORY
```

```python
dataframe.head()
```

```python
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes': grouped_data})
plt.plot(result.index, result['votes'], c="blue", marker="o")
plt.xlabel("Type of restaurant", c="red", size=20)
plt.ylabel("Votes", c="green", size=20)
plt.show()
```

```python
#Conclusion - Dinining resturants has recieved maximum votes
```

```python
dataframe['rate'].value_counts().plot(kind='bar', color="orange")
plt.title("Ratings distribution")
plt.xlabel("Ratings")
plt.ylabel("Count")
plt.show()

```

```python
plt.hist(dataframe['rate'], bins=3, color="green")
plt.title("Ratings distribution")
plt.show()
```

```python
#The majority resturants recieved ratings from 3.5 to 4
```

```python
sns.countplot(x='approx_cost(for two people)', data=dataframe, color='violet')
plt.show()

```

```python
#Which mode gives high rating(Online/offline)?
```

```python
dataframe.head()
```

```python
plt.figure(figsize = (6,6))
sns.boxplot(x = 'online_order' , y = 'rate', data = dataframe)
```

```python
#OFFLINE order recieves lower ratings in comparison to ONLINE orders
```

```python
dataframe.head()
```

```python
pivot_table = dataframe.pivot_table(index='listed_in(type)', columns='online_order' , aggfunc='size' , fill_value=0)
sns.heatmap(pivot_table,annot=True, cmap="YlGnBu" , fmt='d')
plt.title("Heatmap")
plt.xlabel("Online Order")
plt.ylabel("Listed in (Type)")
plt.show()
```

```python
#CONCLUSION - Dining resturants primarily accept offline orders, whereas cafes primarily recieve online orders.
```
